```python
#Creating a function to get the Enrollment for each day
import re

# Extracting day from the Description column
def extract_day(description):
    days = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
    for day in days:
        if day in description:
            return day
    return None

df2["Day"] = df2["Description"].apply(extract_day)
#To get the output in order of days
day_order = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
df2['Day'] = pd.Categorical(df2['Day'],categories=day_order,ordered=True)

# Grouping by Day to see average enrollment rate
grouped_day = df2.groupby("Day")["Enrolled"].mean()

# Plotting average enrollment rate by day
plt.figure(figsize=(10,6))
grouped_day.plot(kind='bar', color='skyblue')
plt.title("Average Enrollment Rate by Day of the Week")
plt.ylabel("Average Enrollment Rate")
plt.xlabel("Day of the Week")
plt.grid(axis='y')
plt.show()
```

```python
season_game_group = df2.groupby(['Season', 'Game'])['Enrolled'].sum().reset_index()

# Now let's create a line plot for each season.
seasons = season_game_group['Season'].unique()

# Set up the matplotlib figure with subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# Adjust the layout
plt.tight_layout(pad=5.0)

# Loop through each season and create a bar plot on its respective subplot
for i, season in enumerate(seasons):
    # Select the current axis
    ax = axes[i//2, i%2]
    
    # Filter the data for the current season
    season_data = season_game_group[season_game_group['Season'] == season]
    
    # Create the bar plot
    ax.bar(season_data['Game'], season_data['Enrolled'], label=season)
    
    # Set the title and labels
    ax.set_title(f'Total Members Enrolled by Each Game on Season - {season}')
    ax.set_xlabel('Game')
    ax.set_ylabel('Enrolled Count')
    ax.legend()
    ax.set_xticklabels(season_data['Game'], rotation=45)
```

```python
# For the new requirement, we want to plot the top four games for each season over the years.
# The dataset should contain the year, the game, the season, and the enrollment counts.
# We will create a line plot for each of the top four games in each season.
# Group the data by 'Year', 'Season', and 'Game' and get the sum of 'Enrolled'
year_season_game_group = df2.groupby([df2['Year'].dt.year, 'Season', 'Game'])['Enrolled'].sum().reset_index()

# Set up the matplotlib figure with subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# Adjust the layout
plt.tight_layout(pad=5.0)

# We will assume the 'top games' are the ones with the highest total enrollment across the dataset
top_games = df2.groupby('Game')['Enrolled'].sum().nlargest(4).index

# Loop through each season and each subplot to create a line plot for the top games
for i, season in enumerate(seasons):
    # Select the current axis
    ax = axes[i//2, i%2]

    # For each top game, create a line plot
    for game in top_games:
        # Filter the data for the current season and game
        game_data = year_season_game_group[(year_season_game_group['Season'] == season) &
                                           (year_season_game_group['Game'] == game)]

        # Create the line plot if there is any data
        if not game_data.empty:
            ax.plot(game_data['Year'], game_data['Enrolled'], marker='o', label=game)

    # Set the title and labels
    ax.set_title(f'Top Four Games in {season}')
    ax.set_xlabel('Year')
    ax.set_ylabel('Enrolled Count')
    ax.legend()
```

```python
# Analyzing data for scheduling

# Grouping data by Season, Day, Sport, and Description to understand frequency and enrollment
grouped_data = df2.groupby(['Season', 'Day', 'Sport', 'Description']).agg({'Enrolled': 'mean', 'Max Count': 'max'}).reset_index()

# Sorting the data based on average enrollment (popularity) and then by day for each season
sorted_data = grouped_data.sort_values(['Season', 'Enrolled', 'Day'], ascending=[True, False, True])

# Extending the schedule creation for all seasons and more activities
# The schedule will prioritize activities based on average enrollment and ensure distribution across the week

# Creating a dictionary to hold the schedule for each season
schedule_dict = {}

# Iterating through each season
for season in sorted_data['Season'].unique():
    season_schedule = sorted_data[sorted_data['Season'] == season]
    
    # Creating a schedule for each day of the week
    for day in ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']:
        day_schedule = season_schedule[season_schedule['Day'] == day].head(2)  # Taking top 2 activities for each day
        
        # Adding to the schedule dictionary
        if season not in schedule_dict:
            schedule_dict[season] = {}
        schedule_dict[season][day] = day_schedule[['Sport', 'Description', 'Enrolled', 'Max Count']].to_dict('records')
    # Dropping NaN values from the schedules

# Function to create and clean the schedule DataFrame for a given season
def create_clean_season_schedule(season, schedule_dict):
    season_schedule_list = []
    for day, activities in schedule_dict[season].items():
        for activity in activities:
            # Skip the activity if it has NaN values or if the Average Enrollment is 0.0
            if pd.isna(activity['Enrolled']) or pd.isna(activity['Max Count']) or activity['Enrolled'] == 0.0:
                continue

            season_schedule_list.append({
                'Day': day,
                'Sport': activity['Sport'],
                'Activity': activity['Description'],
                'Average Enrollment': activity['Enrolled'],
                'Max Capacity': activity['Max Count']
            })
    return pd.DataFrame(season_schedule_list)


# Creating and cleaning schedule DataFrames for each season
season_schedules = {}
for season in schedule_dict.keys():
    season_schedules[season] = create_clean_season_schedule(season, schedule_dict)
```
